## EDA plot 2 - Barplot: Comparing number of each type of tomato (look at distribution)

```python
import os
import matplotlib.pyplot as plt

# Set the base path where the uploaded folders are located
base_path = "/content/sample_data/"

# Define folder names
folder_names = ['ripe', 'unripe', 'old', 'damaged']
folder_counts = []

# Count files in each folder
for folder in folder_names:
    folder_path = os.path.join(base_path, folder)
    if os.path.exists(folder_path):
        count = len([f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))])
    else:
        count = 0  # Folder doesn't exist
    folder_counts.append(count)

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(folder_names, folder_counts, color=['green', 'yellow', 'blue', 'red'])
plt.title('Comparing Count of Each Type of Tomato')
plt.xlabel('Type of Tomato')
plt.ylabel('Number of Tomato Pictures')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## EDA 5: Stacked Bar Chart of Pixel Intensity


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob

damaged = glob.glob('/content/sample_data/damaged/*.jpg')
old = glob.glob('/content/sample_data/old/*.jpg')
unripe = glob.glob('/content/sample_data/unripe/*.jpg')
ripe = glob.glob('/content/sample_data/ripe/*.jpg')


# Helper function to generate grayscale histogram for a folder
def get_grayscale_histogram(image_paths):
    all_pixels = []
    for path in image_paths:
        img = cv2.imread(path)           # Load image
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
        all_pixels.extend(gray.flatten())             # Flatten to 1D array
    hist, _ = np.histogram(all_pixels, bins=np.arange(257))  # 0–255
    return hist

#Histograms for each tomato class
damaged_hist = get_grayscale_histogram(damaged)
old_hist = get_grayscale_histogram(old)
unripe_hist = get_grayscale_histogram(unripe)
ripe_hist = get_grayscale_histogram(ripe)

#Stacked bar chart
bins = np.arange(256)  # X-axis: pixel intensities from 0–255

plt.figure(figsize=(14, 6))
plt.bar(bins, old_hist, label='Old')
plt.bar(bins, ripe_hist, bottom=old_hist, label='Ripe')
plt.bar(bins, unripe_hist, bottom=old_hist + ripe_hist, label='Unripe')
plt.bar(bins, damaged_hist, bottom=old_hist + ripe_hist + unripe_hist, label='Damaged')

plt.xlabel('Pixel Intensity (0–255)')
plt.ylabel('Frequency')
plt.title('Stacked Pixel Intensity Distribution by Tomato Type (Grayscale)')
plt.legend()
plt.tight_layout()
plt.show()
```

# Part 2: CNN Analysis

## Set image size and define train-test split

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator


img_height, img_width = 128, 128  # define image size
batch_size = 32

datagen = ImageDataGenerator(
    rescale=1.0/255,           # normalize pixel values to [0, 1]
    validation_split=0.2       # 80/20 train test split
)

# training data
train_data = datagen.flow_from_directory(
    '/content/sample_data/tomatoes',
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',  # One-hot encode the labels
    subset='training'
)

# validation data
val_data = datagen.flow_from_directory(
    '/content/sample_data/tomatoes',
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)
```
